```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
# every data has been arranged in good format:
```

```python
movies['overview'][0]
#it is a string and we will make it list:
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# we need to attach all the words of particular list for example cast because if we will not do that it will become very 
# difficult for model to diffrentiate between same names example sam in cast and sam mendes in crew.
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
# vectorization to done here (text = vector) vector is something has value between x and y when we will convert every text to vector 
# every movie will become vector now if someone choosed any single movie(vetcor) and according to that movie they will ask  to suggest
# movies then according to closest vector movie will be shown to the end user.
```

```python
# we will remove the stop words(the words which used to sentance formation it dosen't has any use in meaning  example are,and,of)
# by the use of laibrary we will vectorize by using count vectorizer.
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# we have to find distance of every single movie with different movies.we will calculate cosine distance,angle between them 
# if distance is more there is less similarieties.
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1 : 6]
```

```python
sorted(similarity[0],reverse=True)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1 : 6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
